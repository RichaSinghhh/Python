## Seaborn Library

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import scipy as sp
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

## Relational Plot
- Scatter Plot
- Line Plot

## Scatter Plot

```python
#Scatter Plot -> axes level function
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'sex')
```

### Both are same plots

```python
#relplot -> figure level -> square shape
sns.relplot (data = tips, x = 'total_bill', y = 'tip', kind = 'scatter', hue = 'sex', style = 'time',size = 'size' )
```

## Line Plot

```python
gap = px.data.gapminder()
temp_df = gap[gap['country'] == 'India']
temp_df
```

```python
# axes level function
sns.lineplot(data = temp_df, x = 'year', y = 'lifeExp')
```

### Both are Same graph

```python
#using relplot
sns.relplot(data = temp_df, x = 'year', y = 'lifeExp', kind = 'line')
```

- kdeplot

### axes level -> histplot -> kdeplot -> rugplot

```python
# plotting univariate histogram
tips #(data)
```

## kdeplot
- Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate.  

```python
#kdeplot
tips #data
```

```python
sns.kdeplot(data = tips, x = 'total_bill')
```

```python
sns.displot(data = tips, x = 'total_bill', kind = 'kde')
```

```python
# hue -> fill
sns.displot(data = tips, x = 'total_bill', kind = 'kde', hue = 'sex', fill = True)
```

```python
# hue -> fill
plt.figure(figsize = (15,7))#it's a figure level function, that's why it's not working
sns.displot(data= tips, x = 'total_bill', kind = 'kde', hue = 'sex', fill = True)
```

```python
#this function is used for figure level function,for manipulate height & width. 
sns.displot(data= tips, x = 'total_bill', kind = 'kde', hue = 'sex', fill = True, height = 10, aspect = 2)
```

## Rugplot
- Plot marginal distributions by drawing ticks along the x and y axes.
- This function is intended to complement other plots by showing the location of individual observation in an unobtrusive way.

```python
sns.kdeplot(data = tips, x = 'total_bill')
sns.rugplot(data = tips, x = 'total_bill')
```

## Bivariate histogram
- A bivariate histogram bins the data within rectangles that tile the plot
- and then shows the count of observations within each rectangle with the fill color

```python
sns.histplot(data = tips, x = 'total_bill', y = 'tip') #normal 2D histplot
```

```python
sns.displot(data = tips, x = 'total_bill', y = 'tip', kind = 'hist')
```

## Bivariate kdeplot
- a bivariate KDE plot smooths the (x, y) observation with a 2D Gussain

```python
sns.kdeplot(data = tips, x = 'total_bill', y = 'tip')
```

## Matrix Plot

### Heatmap
- plot rectangular data as a color-encoded matrix

```python
gap #data
```

```python
temp_df = gap.pivot(index = 'country', columns = 'year', values = 'lifeExp')

#heatmap is a axes level function
sns.heatmap(temp_df)
```

```python
plt.figure(figsize = (15,15))
sns.heatmap(temp_df)
```

```python

temp_df = gap[gap ['continent'] == 'Europe'].pivot(index = 'country', columns = 'year', values = 'lifeExp')

plt.figure(figsize = (15,15))
sns.heatmap(temp_df)
```

```python
#annot
temp_df = gap[gap ['continent'] == 'Europe'].pivot(index = 'country', columns = 'year', values = 'lifeExp')

plt.figure(figsize = (15,15))
sns.heatmap(temp_df, annot = True,linewidth = 0.5) #annot is used for numbering
 #linewidth is used to give some space
```

```python
plt.figure(figsize = (15,15)) #cmap is used for coloring
sns.heatmap(temp_df, annot = True, linewidth = 0.5, cmap = 'viridis') #cmap###########3####sdhshdsldhsfhfshjfljeshhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhh= summer,winter,rainny,viridis
```
