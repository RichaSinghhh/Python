## Seaborn Library

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

## Relational Plot
- Scatter Plot
- Line Plot

## Scatter Plot

```python
#Scatter Plot -> axes level function
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'sex')
```

### Both are same plots

```python
#relplot -> figure level -> square shape
sns.relplot (data = tips, x = 'total_bill', y = 'tip', kind = 'scatter', hue = 'sex', style = 'time',size = 'size' )
```

## Line Plot

```python
gap = px.data.gapminder()
temp_df = gap[gap['country'] == 'India']
temp_df
```

```python
# axes level function
sns.lineplot(data = temp_df, x = 'year', y = 'lifeExp')
```

### Both are Same graph

```python
#using relplot
sns.relplot(data = temp_df, x = 'year', y = 'lifeExp', kind = 'line')
```

```python
# hue -> style
temp_df = gap[gap['country'].isin(['India','Pakistan','China'])]
temp_df
```

```python
#using relplot
sns.relplot(kind = 'line', data = temp_df, x = 'year', y = 'lifeExp', hue = 'country')
```

### Both are same graph

```python
sns.lineplot(data = temp_df, x = 'year', y = 'lifeExp', hue = 'country')
```

```python
# hue -> style
temp_df = gap[gap['country'].isin(['India','Brazil','Germany'])]
temp_df
```

```python
#using relplot
sns.relplot(kind = 'line', data = temp_df, x = 'year', y = 'lifeExp', hue = 'country', style = 'continent',size = 'continent')
```

## Facet Plot

```python
# facet plot
tips
```

```python
# facet plot(in row format)
sns.relplot(data = tips, x = 'total_bill', y = 'tip', kind = 'scatter', row = 'sex')
```

```python
# facet plot (in column format)
sns.relplot(data = tips, x = 'total_bill', y = 'tip', kind = 'scatter', col = 'sex', hue = 'sex')
```

```python
# facet plot (row & column format)
sns.relplot(data = tips, x = 'total_bill', y = 'tip', kind = 'scatter', row = 'day', col = 'sex')
```

```python
# facet plot -> figure level function -> work with relplot
# it will not work with scatterplot and lineplot
sns.relplot(data = tips, x = 'total_bill', y = 'tip', kind = 'line', row = 'day', col = 'sex')
```

## 2. Distribution Plots
- used for univariate analysis
- used to find out the distribution
- Range of the observation
- Central Tendency
- is the data bimodal?
- Are there outliers?

# Plots under distribution plot
- histplot
- kdeplot
- rugplot

### figure level -> displot
### axes level -> histplot -> kdeplot -> rugplot

### Histogram

```python
# plotting univariate histogram
tips #(data)
```

```python
# plotting univariate histogram
sns.histplot(data = tips, x = 'total_bill')
```

```python
sns.displot(data = tips, x = 'total_bill', kind = 'hist')
```

```python
#use of bins parameter
sns.displot(data = tips, x = 'total_bill', kind = 'hist', bins = 20)
```

```python
#bins parameter
sns.displot(data = tips, x = 'total_bill', kind = 'hist', bins = 2)
```
