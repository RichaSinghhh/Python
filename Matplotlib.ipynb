### Linear Graph

```python
x = [1,2,3,4]
y = [5,6,7,8]
plt.plot(x,y)
```

```python
x = [1,2,3,4]
y = [5,6,7,8]
c = 'r'
plt.plot(x,y,c)
plt.show()
```

```python
x = [5,2,9,4,7]
y = [10,5,8,4,2]

plt.plot(x,y)
plt.show()
```

## Marker

```python
x = [5,2,9,4,7]
y = [10,5,8,4,2]

plt.plot(x,y,marker = 'o')
plt.show()
```

```python
#in this graph x,y axis both are seperated
x = [5,2,9,4,7]
y = [10,5,8,4,2]

plt.plot(x,y)
plt.plot(y,marker = '>') #use this symbol(>)for 
plt.show()
```
