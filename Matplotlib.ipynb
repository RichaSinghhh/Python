### Linear Graph

```python
x = [1,2,3,4]
y = [5,6,7,8]
plt.plot(x,y)
```

```python
x = [1,2,3,4]
y = [5,6,7,8]
c = 'r'
plt.plot(x,y,c)
plt.show()
```

```python
x = [5,2,9,4,7]
y = [10,5,8,4,2]

plt.plot(x,y)
plt.show()
```

## Marker

```python
x = [5,2,9,4,7]
y = [10,5,8,4,2]

plt.plot(x,y,marker = 'o')
plt.show()
```

```python
#in this graph x,y axis both are seperated
x = [5,2,9,4,7]
y = [10,5,8,4,2]

plt.plot(x,y) #blue line is x axis,yellow line is y axis
plt.plot(y,marker = '>') #use this symbol(>)for tringale
plt.plot(x,marker = 'o')
plt.show()
```

## Bar Graph

```python
x = [1,2,3,4]
y = [5,6,7,8]

plt.bar(x,y)
plt.show()
```

```python
#how to change bar graph color
x = [1,2,3,4]
y = [5,6,7,8]
c = ['r','g','b','y']
plt.bar(x,y, color = c)
plt.show()
```

## Scatter Plot

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [4,2,3,5,1,8]

plt.scatter(x,y)
plt.show()
```

```python
#how to give title in the chart
x = [1,2,3,4,5,6]
y = [4,2,3,5,1,8]
plt.title("Scatter Plot")
plt.scatter(x,y)
plt.show()
```

```python
#how to give label in the chart
x = [1,2,3,4,5,6]
y = [4,2,3,5,1,8]
plt.title("Scatter Plot")
plt.xlabel("month")
plt.ylabel("numbers")
plt.scatter(x,y)
plt.show()
```

```python
#how to change color
x = [1,2,3,4,5,6]
y = [4,2,3,5,1,8]
plt.title("Scatter Plot")
plt.xlabel("month")
plt.ylabel("numbers")
plt.scatter(x,y,color = 'g')
plt.show()
```

```python
#I want to change differant differant colors in scatter plot
x = [1,2,3,4,5,6]
y = [4,2,3,5,1,8]
plt.title("Scatter Plot")
plt.xlabel("month")
plt.ylabel("numbers")
c = ['red','magenta','green','orange','blue','brown']
plt.scatter(x,y, color = c)
plt.show()
```

```python
#how to resize our scatter plot
x = [1,2,3,4,5,6]
y = [4,2,3,5,1,8]
plt.title("Scatter Plot")
plt.xlabel("month")
plt.ylabel("numbers")
c = ['red','magenta','green','orange','blue','brown']
plt.scatter(x,y, color = c ,s=200 )
plt.show()
```
