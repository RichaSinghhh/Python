### How to delete null/NaN values in rows

```python
df1.dropna() #Data is not permanently deleted with this function.
```

```python
df1
```

```python
#Data is permanently deleted with this function.
df1.dropna(inplace=True)
```

### How to delete null/NaN values in column

```python
df2 = sns.load_dataset('titanic')
df2
```

```python
df2.dropna(axis = 1) #column is called axis(Data is not permanently deleted with this function.)
```

```python
df2
```

```python
df2.dropna(axis = 1,inplace=True)#Data is permanently deleted with this function.
df2
```

### In place of NaN, give a new value according to your choice as follows

```python
df3 = sns.load_dataset('titanic')
df3
```

```python
df3.fillna("richa") #Data is not permanently deleted with this function.
```

```python
df3
```

```python
df3.fillna("richa",inplace=True) #Data is permanently deleted with this function.
```
