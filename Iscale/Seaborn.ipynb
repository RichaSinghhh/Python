## Seaborn Library

### Seaborn is a library for making statistical graph in python

```python
import seaborn as sns #(sns is a allias)
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = [1,2,3,4,5,6,7,8]
var1 =[8,7,6,5,4,3,2,1]
```

### made graph through matplotlib

```python
plt.plot(var,var1)
plt.show()
```

## Line graph in seaborn library

```python
sns.lineplot(x = var, y = var1)
plt.show()
```

```python
var = [1,2,3,4,5,6,7,8]
var1 =[8,7,6,5,4,3,2,1]

import pandas as pd

df = pd.DataFrame({"var":var,"var1":var1})
sns.lineplot(x= 'var', y = 'var1',data = df)
plt.show()
```

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/refs/heads/master/penguins.csv"
df1 = pd.read_csv(url)
df1
```

```python
sns.lineplot(x = 'bill_length_mm', y = 'flipper_length_mm', data = df1) #show line graph
plt.show()
```

```python
#divide in terms of gender by catagorical term (hue is used for catagories data.)
sns.lineplot(x = 'bill_length_mm', y = 'flipper_length_mm', data = df1 , hue = "sex") #show line graph
plt.show()
```

```python
sns.lineplot(x = 'bill_length_mm', y = 'flipper_length_mm', data = df1 , hue = "sex", palette = "rocket_r" ) #palette is used for coloring
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm", y = "flipper_length_mm",data = df1 , hue = 'sex', style = 'sex', palette = 'rocket_r', markers = ['o','>'])
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm", y = "flipper_length_mm",data = df1 , hue = 'sex', style = 'sex', palette = 'rocket_r', markers = ['o','>'])
plt.grid() #grid is used for background lines it's known as grid.
plt.show() 
```

```python
sns.lineplot(x = "bill_length_mm", y = "flipper_length_mm",data = df1 , hue = 'sex', style = 'sex', palette = 'rocket_r', markers = ['o','>'])
plt.title("this is a line graph related to penguins")#title is used to give the title of chart
plt.show() 
```

## Order Parameter

```python
order1 = ['Dream','Torgersen','Biscoe']
sns.barplot(x ='island', y = 'bill_length_mm',data = df1, hue = 'sex', order = order1)
plt.show()
```

## hue_order

```python
order1 = ['Dream','Torgersen','Biscoe']
sns.barplot(x = 'island', y = 'bill_length_mm', data = df1, hue = 'sex', order = order1, hue_order= ['Female', 'Male'])
plt.show()
```

```python
order1 = ['Dream', 'Torgerson','Biscoe']
sns.barplot(x = 'island', y = 'bill_length_mm',data = df1, color = 'yellow')
plt.show()
```

```python
sns.barplot(x = 'island', y = 'bill_length_mm',data = df1, hue = 'sex', palette = 'prism')
plt.show()
```

```python
sns.barplot(x = 'island', y = 'bill_length_mm',data = df1, hue = 'sex', palette = 'gist_heat_r')
plt.show()
```

## Saturation

```python
sns.barplot(x = 'island', y = 'bill_length_mm',data = df1, hue = 'sex', palette = 'gist_heat_r', saturation = 0.2)
plt.show() #saturation is used for fad colors in graph
```
